```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loss_raw = pd.read_csv('E:/my-whitecaps/code/organized/model/test/raw_test/training_statistics_test.csv')
loss_vgg = pd.read_csv('E:/my-whitecaps/code/organized/model/test/vgg_test/training_statistics_test.csv')
loss_resnet = pd.read_csv('E:/my-whitecaps/code/organized/model/test/resnet50_test/training_statistics_test.csv')
```

```python
lr_rate_list_raw = loss_raw['lr_rate_list'].values
lr_rate_list_vgg = loss_vgg['lr_rate_list'].values
lr_rate_list_res = loss_resnet['lr_rate_list'].values
```

```python
from ast import literal_eval
plt.figure(figsize=(8,8))
plt.axes(yscale="log")
plt.plot([literal_eval(i) for i in lr_rate_list_raw],linestyle='-',linewidth = 3,color='k')
plt.plot([literal_eval(i) for i in lr_rate_list_vgg],linestyle='-.',linewidth = 3,color='k')
plt.plot([literal_eval(i) for i in lr_rate_list_res],linestyle='--',linewidth = 3,color='k')

font = {'family' : 'Times New Roman', 
        'weight' : 'normal', 
        'size' : 32}
plt.xlabel('Epoch', font)
plt.ylabel('learing rate during training', font)
plt.xticks(fontproperties='Times New Roman', size = 32)
plt.yticks(fontproperties='Times New Roman', size = 32)
plt.grid()
font_legend1 = {'family' : 'Times New Roman', 
        'weight' : 'normal', 
        'size' : 20}
plt.legend(['UNet Raw','VGG16','Resnet50'],prop = font_legend1)
plt.savefig('learn_rat_compare.png',bbox_inches='tight')
```

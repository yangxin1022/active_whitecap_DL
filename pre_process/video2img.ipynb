```python
import os
from pathlib import Path


'''
The following code can loop over all the GoPro data folders to convert 
MP4 video record into images at a fixed FPS. Here we want one image each
30 seconds, FPS = 1/30
'''

day_video = []
for fname in os.listdir('F:/'):
    if fname[-5:] == 'Gopro':
        day_video.append(fname)
print('The list of GoPro record date:\n'+ '\n'.join(day_video))

loc_camera = ['left', 'behind', 'right']

for dv in day_video:
    for lcc in loc_camera:
        video_path = Path(f'F:/{dv}/{lcc}')
        if video_path.exists():
            video_num = [fn for fn in os.listdir(f'F:/{dv}/{lcc}')
             if fn.endswith('MP4')]
        for vn in video_num:
            video_path = Path(f'F:/{dv}/{lcc}/{vn}')
            if video_path.exists():
                video_cap = cv2.VideoCapture(f'F:/{dv}/{lcc}/{vn}')
                fps = video_cap.get(cv2.CAP_PROP_FPS)
                frames = video_cap.get(cv2.CAP_PROP_FRAME_COUNT)
                print("fps=",int(fps),"Frames=",int(frames))
                video_cap.set(cv2.CAP_PROP_FPS, 1/30)
            
                sec = 0
                count = 0
                frame_rate = 30 # one image every 30 seconds
                success = True
                while success:
                    # CAP_PROP_POS_MSEC: current position of the video file in milliseconds
                    video_cap.set(cv2.CAP_PROP_POS_MSEC, sec*1000)
                    success, image = video_cap.read()
                    if not success:
                        continue
                    cv2.imwrite(f'E:/GoPro_training/{dv}_{lcc}_{vn[0:-4]}_frame_{count}.png', image)
                   
                    print('Read a new frame: ', success)
                    count += 1
                    sec = sec + frame_rate
                video_cap.release()
```

```python
'''
Randomly pick 2000 images
Crop the images and save them
'''
import numpy as np

# set the seed for random number
rd = np.random.RandomState(1)
# create a 1*2000 vector whose elements range from 0 to 9893, which are the
# number of the images.
vector = rd.randint(0, 9893, (1, 2000))

import os
import numpy as np

raw_img_list = os.listdir("E:/GoPro_training/raw")
raw_img_list = np.array(raw_img_list)
training_1k = raw_img_list[vector]
training_1k = np.squeeze(training_1k)

for f in training_1k:
    img = cv2.imread(f"E:/GoPro_training/raw/{f}")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_crop = gray[540:, 0:]
    left_part = gray_crop[:, 0:960]
    right_part = gray_crop[:, 960:]
    
    cv2.imwrite(f"E:/GoPro_training/cropped_4k/{f[0:-4]}"+"left.png", left_part)
    cv2.imwrite(f"E:/GoPro_training/cropped_4k/{f[0:-4]}"+"right.png", right_part)
```

```python
import cv2
import os

def extract_frames(video_path, output_dir):
    # Open the video file
    video = cv2.VideoCapture(video_path)
    
    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Initialize variables
    frame_count = 0
    success = True
    
    # Extract frames from the video
    while success:
        # Read the next frame
        success, frame = video.read()
        
        # If the frame was read successfully
        if success:
            # Save the frame as an image
            frame_path = os.path.join(output_dir, f"frame_{frame_count}.jpg")
            cv2.imwrite(frame_path, frame)
            
            frame_count += 1
    
    # Release the video file
    video.release()

# Specify the path to the video file
video_path = 'path/to/video.mp4'

# Specify the output directory for the frames
output_dir = 'path/to/output_directory'

# Extract frames from the video
extract_frames(video_path, output_dir)
```
